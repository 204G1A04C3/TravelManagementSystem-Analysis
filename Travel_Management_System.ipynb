# Visualizing the data

```python
plt.figure(figsize=(10, 6))  # Set the figure size
plt.bar(NumCustomers_data['Address'], NumCustomers_data['NumCustomers'])  # Create a bar plot
plt.xlabel('Address')  # Set the label for the x-axis
plt.ylabel('Number of Customers')  # Set the label for the y-axis
plt.title('Number of Customers per Address')  # Set the title of the plot
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()  # Show the plot
```

```python
# Define status_colors dictionary to specify colors for each booking status
status_colors = {
    'Confirmed': 'green',    # Green color for confirmed bookings
    'Pending': 'orange',     # Orange color for pending bookings
    'Cancelled': 'red'       # Red color for cancelled bookings
}

# Create subplots with a row for each status
fig, axs = plt.subplots(len(status_colors), 1, figsize=(10, 8), sharex=True)

# Plot each status separately
for i, (status, color) in enumerate(status_colors.items()):
    # Filter data for the current status
    data = Bookings_df[Bookings_df['Status'] == status]
    # Plot the booking trend over time for the current status
    axs[i].plot(data['BookingDate'], data['BookingID'], marker='o', linestyle='-', label=status, color=color)
    axs[i].set_ylabel('Booking ID')  # Set y-axis label for the current subplot
    axs[i].set_title(f'Booking Trends Over Time - {status}')  # Set title for the current subplot
    axs[i].grid(True)  # Add gridlines to the current subplot
    axs[i].legend()  # Add legend for the current subplot

# Set common x-axis label and rotate tick labels
plt.xlabel('Booking Date')  # Set x-axis label for the entire figure
plt.xticks(rotation=45)  # Rotate x-axis tick labels for better readability

# Adjust layout to prevent overlap and display the plot
plt.tight_layout()
plt.show()

```

# Visualize the distribution of payment methods using a pie chart 

```python
# Count occurrences of each payment method
payment_mode_counts = Payments_df['PaymentMethod'].value_counts()

# Find the mode (payment method with the highest count)
mode_of_payment = payment_mode_counts.idxmax()

print(f"The mode of payment is '{mode_of_payment}'.")    # Print the mode of payment

# Plotting the distribution of payment methods using a pie chart
plt.figure(figsize=(8, 6))
payment_mode_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title('Payment Method Distribution')
plt.show()
```

- The Payment Method Distribution graph depicts the distribution of payment methods utilized within the Travel management system.

- Payment Methods: The pie chart segments represent different payment methods used by customers, such as Credit Card, PayPal, Cash, Bank Transfer.

- Percentage Breakdown: Each segment of the pie chart displays the proportion of payments made through a specific payment method, indicated as a percentage of the total.

# Visualize the Average Price of Travel Packages by Duration  

```python
import numpy as np 
# Define the bin edges for duration
bin_edges = np.arange(0, TravelPackages_df['DurationDays'].max() + 10, 10)  # Create bin edges for durations with a step of 10 days

# Bin the data by duration and calculate the mean price for each bin
bin_labels = pd.cut(TravelPackages_df['DurationDays'], bins=bin_edges, right=False)  # Bin durations into intervals
price_means = TravelPackages_df.groupby(bin_labels)['Price'].mean()  # Calculate the mean price for each duration bin

# Plot a bar graph
plt.figure(figsize=(10, 6))  # Set the figure size
price_means.plot(kind='bar', color='skyblue')  # Create a bar plot of average prices for each duration bin
plt.xlabel('Duration (Days)')  # Set the label for the x-axis
plt.ylabel('Average Price')  # Set the label for the y-axis
plt.title('Average Price of Travel Packages by Duration')  # Set the title of the plot
plt.xticks(rotation=45)  # Rotate the x-axis labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add gridlines to the y-axis
plt.show()  # Display the plot
```

 # Visualize the Booking trends Over Time by Using Line Chart 

```python
# Convert 'BookingDate' column to datetime format
Bookings_df['BookingDate'] = pd.to_datetime(Bookings_df['BookingDate'])

# Assuming 'BookingDate' is now in datetime format, you can proceed with the rest of the code
Bookings_df['BookingMonth'] = Bookings_df['BookingDate'].dt.to_period('M')  # Extract month from booking date
bookings_per_month = Bookings_df.groupby('BookingMonth')['BookingID'].count()  # Group bookings by month and count the number of bookings

# Plot a line chart to visualize the booking trends over time
plt.figure(figsize=(10, 6))  # Set the figure size
bookings_per_month.plot(kind='line', marker='o')  # Create a line plot of bookings over time
plt.xlabel('Booking Month')  # Set the label for the x-axis
plt.ylabel('Number of Bookings')  # Set the label for the y-axis
plt.title('Booking Trends Over Time')  # Set the title of the plot
plt.xticks(rotation=45)  # Rotate the x-axis labels for better readability
plt.grid(True)  # Add gridlines to the plot
plt.show()  # Display the plot

```
